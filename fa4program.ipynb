Problem 2

```python
correct_password = "colab"
max_attempts = 3
attempts = 0

while True:
    password = input("Enter the password: ")
    attempts += 1

    if password == correct_password:
        print("Access granted! Welcome.")
        break
    else:
        print("Incorrect password. Please try again.")
        if attempts >= max_attempts:
            print(f"You have reached the maximum number of attempts ({max_attempts}). Access denied.")
            break
```
